### –ü—Ä–æ–≤–µ—Ä–∏–º, —á—Ç–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ

–î–ª—è —ç—Ç–æ–≥–æ –≤–≤–µ–¥—ë–º —Ñ–æ—Ä–º—É–ª—É:



–≥–¥–µ:
- `Recovery` ‚Äî —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –æ–±–æ–≥–∞—â–µ–Ω–∏—è;
- `C` ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
- `F` ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ/–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –¥–æ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
- `T` ‚Äî –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –æ—Ç–≤–∞–ª—å–Ω—ã—Ö —Ö–≤–æ—Å—Ç–∞—Ö –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏.

–î–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ –¥–æ–ª—é –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–∞—Ö –∏ —Ö–≤–æ—Å—Ç–∞—Ö. –í–∞–∂–µ–Ω –Ω–µ —Ç–æ–ª—å–∫–æ —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –ø—Ä–æ–¥—É–∫—Ç, –Ω–æ –∏ —á–µ—Ä–Ω–æ–≤–æ–π –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç.

–í—Ä—É—á–Ω—É—é –≤—ã—á–∏—Å–ª–∏–º —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–∞ `rougher.output.recovery`.

```python
numerator = (df_train['rougher.output.concentrate_au'] * 
             (df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au']) )
denominator = (df_train['rougher.input.feed_au'] * 
             (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au']) )
df_train['recovery'] = numerator / denominator * 100
df_train['recovery']
```

–ù–∞–π–¥—ë–º MAE –º–µ–∂–¥—É –Ω–∞—à–∏–º–∏ —Ä–∞—Å—á—ë—Ç–∞–º–∏ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∞:


```python
MAE = (df_train['rougher.output.recovery'] - df_train['recovery']).sum() / df_train.shape[0]
print('recovery_–ú–ê–ï, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω–æ–µ –≤—Ä—É—á–Ω—É—é:', MAE)
print('recovery_–ú–ê–ï, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω–æ–µ —Ñ—É–Ω–∫—Ü–∏–µ–π:', mean_absolute_error(df_train['recovery'], df_train['rougher.output.recovery']))
```

## –ê–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –º–µ–Ω—è–µ—Ç—Å—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –º–µ—Ç–∞–ª–ª–æ–≤ (Au, Ag, Pb) –Ω–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö –æ—á–∏—Å—Ç–∫–∏

#### –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç —Ä–∏—Å–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–µ—â–µ—Å—Ç–≤–∞.

```python
def concentration_of_metals(metal, legend):
    # —Ä–∞–∑–º–µ—Ä –≥—Ä–∞—Ñ–∏–∫–∞
    plt.figure(figsize=(15,5))
    
    # –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞
    df_full[metal[1]].hist(bins=25, label='rougher.output.feed', alpha=0.5) # —Ñ–ª–æ—Ç–∞—Ü–∏—è —Å—ã—Ä—å—ë
    df_full[metal[2]].hist(bins=25, label='rougher.output.concentrate', alpha=0.5) # —Ñ–ª–æ—Ç–∞—Ü–∏—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç
    df_full[metal[3]].hist(bins=25, label='primary_cleaner.output', alpha=0.5) # –ø–µ—Ä–≤—ã–π —ç—Ç–∞–ø
    df_full[metal[4]].hist(bins=25, label='secondary_cleaner.output', alpha=0.5) # –≤—Ç–æ—Ä–æ–π —ç—Ç–∞–ø
    plt.title('–î–∏–∞–≥—Ä–∞–º–º–∞ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ –º–µ—Ç–∞–ª–ª–∞ ' + metal[0])
    
    # –ø–æ–¥–ø–∏—Å–∏
    plt.legend(loc=legend)
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è')
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ')
    
    plt.show()
    
    # –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –º–µ—Ç–∞–ª–ª–∞ –≤ —Å—ã—Ä—å–µ
    print('–ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –º–µ—Ç–∞–ª–ª–∞ –≤ —Å—ã—Ä—å–µ:\n', df_full[[metal[2], 
                                                      metal[3], 
                                                      metal[4]]].mean())
    # –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –º–µ—Ç–∞–ª–ª–∞ –≤ —Ö–≤–æ—Å—Ç–∞—Ö
    print('\n–ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –º–µ—Ç–∞–ª–ª–∞ –≤ —Ö–≤–æ—Å—Ç–∞—Ö: \n', df_full[[metal[5], metal[6], metal[7]]].mean())
```

#### –ó–æ–ª–æ—Ç–æ

```python
Au = ['–∑–æ–ª–æ—Ç–æ (Au)',  # –Ω–∞–∑–≤–∞–Ω–∏–µ
      'rougher.input.feed_au', 'rougher.output.concentrate_au', # —Å—ã—Ä—å—ë
      'primary_cleaner.output.concentrate_au', 'final.output.concentrate_au', # —Å—ã—Ä—å—ë
      'rougher.output.tail_au', 'primary_cleaner.output.tail_au', 'secondary_cleaner.output.tail_au'] # —Ö–≤–æ—Å—Ç—ã

# –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ –∑–æ–ª–æ—Ç–∞
concentration_of_metals(Au, 'upper left')
```

#### –°–µ—Ä–µ–±—Ä–æ

```python
Ag = ['—Å–µ—Ä–µ–±—Ä–æ (Ag)',  # –Ω–∞–∑–≤–∞–Ω–∏–µ
      'rougher.input.feed_ag', 'rougher.output.concentrate_ag', # —Å—ã—Ä—å—ë
      'primary_cleaner.output.concentrate_ag', 'final.output.concentrate_ag', # —Å—ã—Ä—å—ë
      'rougher.output.tail_ag', 'primary_cleaner.output.tail_ag', 'secondary_cleaner.output.tail_ag'] # —Ö–≤–æ—Å—Ç—ã

# –∏–∑–º–µ–Ω–µ–Ω–∏—è –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ —Å–µ—Ä–µ–±—Ä–∞
concentration_of_metals(Ag, 'upper right')
```

### –°—Ä–∞–≤–Ω–∏–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–æ–≤ –≥—Ä–∞–Ω—É–ª —Å—ã—Ä—å—è –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö

–î–ª—è —ç—Ç–æ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–µ–º –≥—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ `kdeplot()` –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ **seaborn**. \
–ö–æ–¥ –æ—Ç—Ä–∏—Å–æ–≤–∫–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ç–∞–∫–∂–µ –ø–æ–º–µ—Å—Ç–∏–º –≤ —Ñ—É–Ω–∫—Ü–∏—é.

```python
def granule_density(stage, limits):
    # —Ä–∞–∑–º–µ—Ä –≥—Ä–∞—Ñ–∏–∫–∞
    plt.figure(figsize=(15, 5))
    
    # –≥—Ä–∞—Ñ–∏–∫
    sns.kdeplot(df_train[stage], 
            label=f'train {stage}', alpha=0.5, shade=True)
    sns.kdeplot(df_test[stage], 
            label=f'test {stage}', alpha=0.7, shade=True)
    
    # –¥–æ–ø. –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
    plt.xlim(limits)
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è')
    plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã')
    plt.legend(loc='upper right')
    
    plt.show()
    
    # —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π
    print('df_train:')
    print(df_train[stage].describe())
    
    print('\ndf_test:')
    print(df_test[stage].describe())
```

```python
# —Ä–∞–∑–º–µ—Ä –≥—Ä–∞–Ω—É–ª —Å—ã—Ä—å—è –Ω–∞ –≤—Ö–æ–¥–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏
granule_density('rougher.input.feed_size', [0, 175])
```

### –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE

–î–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –≤–≤–µ–¥—ë–º –Ω–æ–≤—É—é –º–µ—Ç—Ä–∏–∫—É –∫–∞—á–µ—Å—Ç–≤–∞ ‚Äî **sMAPE** (–∞–Ω–≥–ª. Symmetric Mean Absolute Percentage Error, ¬´—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ¬ª). \
–û–Ω–∞ –ø–æ—Ö–æ–∂–∞ –Ω–∞ MAE, –Ω–æ –≤—ã—Ä–∞–∂–∞–µ—Ç—Å—è –Ω–µ –≤ –∞–±—Å–æ–ª—é—Ç–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω–∞—Ö, –∞ –≤ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö. –ü–æ—á–µ–º—É —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è? –û–Ω–∞ –æ–¥–∏–Ω–∞–∫–æ–≤–æ —É—á–∏—Ç—ã–≤–∞–µ—Ç –º–∞—Å—à—Ç–∞–± –∏ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞, –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è.

–ú–µ—Ç—Ä–∏–∫–∞ sMAPE –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Ç–∞–∫:


```python
# sMAPE –¥–ª—è –æ–¥–Ω–æ–≥–æ —ç—Ç–∞–ø–∞
def sMAPE(target, pred):
    
    smape = abs(target - pred) / ((abs(target) + abs(pred)) / 2) * 100
    
    #smape = smape.fillna(value=0)
    
    # —Å–∫–ª–∞–¥—ã–≤–∞–µ–º –∏ –¥–µ–ª–∏–º –Ω–∞ N
    smape = np.mean(smape)
    
    return smape
```

```python
# –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Ç–æ–≥–æ–≤–æ–π sMAPE, –∫–æ–≥–¥–∞ –Ω–∞ –≤—Ö–æ–¥–µ target_rouger, pred_rougher, target_final, pred_final
def summary_smape_func(target_rouger, pred_rougher, target_final, pred_final):
    
    # –≤—ã—á–∏—Å–ª–∏–º smape –¥–ª—è rouger
    smape_rouger = sMAPE(target_rouger, pred_rougher)
    
    # –≤—ã—á–∏—Å–ª–∏–º smape –¥–ª—è final
    smape_final = sMAPE(target_final, pred_final)
    
    #  –≤—ã—á–∏—Å–ª–∏–º –∏—Ç–æ–≥–æ–≤—ã–π smape
    summary_smape = 0.25 * smape_rouger + 0.75 * smape_final
    
    return summary_smape
```

```python
# –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏—Ç–æ–≥–æ–≤–æ–π sMAPE, –∫–æ–≥–¥–∞ –Ω–∞ –≤—Ö–æ–¥–µ smape_rouger –∏ smape_final
def summary_smape(smape_rouger, smape_final):
    
    #  –≤—ã—á–∏—Å–ª–∏–º –∏—Ç–æ–≥–æ–≤—ã–π smape
    summary_smape = 0.25 * smape_rouger + 0.75 * smape_final
    
    return summary_smape
```

–°–æ–∑–¥–∞–¥–∏–º –æ–±—ä–µ–∫—Ç –¥–ª—è –æ—Ü–µ–Ω–∫–∏:

```python
smape_scorer = make_scorer(sMAPE, greater_is_better=False)
```

### –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

```python
# –∏–Ω–∏—Ü–∏–∏—Ä—É–µ–º –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model_regression_rougher = LinearRegression()

# –æ—Ü–µ–Ω–∏–º –º–æ–¥–µ–ª—å —Å –ø–æ–º–æ—â—å—é –∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏–∏ features_train, target_train. –í—ã—á–∏—Å–ª–∏–º sMape_rougher
scores_line_reg_rougher = cross_val_score(model_regression_rougher,  
                                          features_train, 
                                          target_train['rougher.output.recovery'], 
                                          scoring = smape_scorer,
                                          cv = 5)
smape_rougher = np.mean(abs(scores_line_reg_rougher))
print('–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ sMAPE_rougher:', round(smape_rougher, 4))
```

```python
# –æ—Ü–µ–Ω–∏–º –º–æ–¥–µ–ª—å —Å –ø–æ–º–æ—â—å—é –∫—Ä–æ—Å—Å–≤–∞–ª–∏–¥–∞—Ü–∏–∏ features_train, target_train. –í—ã—á–∏—Å–ª–∏–º sMAPE_final
scores_line_reg_rougher = cross_val_score(model_regression_rougher,  
                                          features_train, 
                                          target_train['final.output.recovery'], 
                                          scoring = smape_scorer,
                                          cv = 5)
smape_final = np.mean(abs(scores_line_reg_rougher))
print('–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ sMAPE_final:', round(smape_final, 4))
```

```python
# –≤—ã—á–∏—Å–ª–∏–º –∏—Ç–æ–≥–æ–≤—É—é –æ—Ü–µ–Ω–∫—É sMape
smape_total = summary_smape(smape_rougher, smape_final)
print('–°—Ä–µ–¥–Ω—è—è –∏—Ç–æ–≥–æ–≤–∞—è –æ—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ sMAPE:', round(smape_total, 4))
```

### –î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

–ü—Ä–æ–≤–µ–¥–µ–º –æ—Ü–µ–Ω–∫—É –º–æ–¥–µ–ª–∏ –¥–µ—Ä–µ–≤–∞ —Ä–µ—à–µ–Ω–∏–π –Ω–æ —É–∂–µ —Å –ø–æ–¥–±–æ—Ä–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –≥–ª—É–±–∏–Ω—ã.
