### Проверим, что эффективность обогащения рассчитана правильно

Для этого введём формулу:



где:
- `Recovery` — эффективность обогащения;
- `C` — доля золота в концентрате после флотации/очистки;
- `F` — доля золота в сырье/концентрате до флотации/очистки;
- `T` — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Важен не только финальный продукт, но и черновой концентрат.

Вручную вычислим эффективность на обучающей выборке для признака `rougher.output.recovery`.

```python
numerator = (df_train['rougher.output.concentrate_au'] * 
             (df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au']) )
denominator = (df_train['rougher.input.feed_au'] * 
             (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au']) )
df_train['recovery'] = numerator / denominator * 100
df_train['recovery']
```

Найдём MAE между нашими расчётами и значением признака:


```python
MAE = (df_train['rougher.output.recovery'] - df_train['recovery']).sum() / df_train.shape[0]
print('recovery_МАЕ, рассчитанное вручную:', MAE)
print('recovery_МАЕ, рассчитанное функцией:', mean_absolute_error(df_train['recovery'], df_train['rougher.output.recovery']))
```

Средняя абсолютная ошибка **близка к нулю в обоих случаях**, значит признак `recovery` в таблице посчитан правильно.

### Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки

#### Напишем функцию, которая будет рисовать график для каждого вещества.

```python
def concentration_of_metals(metal, legend):
    # размер графика
    plt.figure(figsize=(15,5))
    
    # гистограмма
    df_full[metal[1]].hist(bins=25, label='rougher.output.feed', alpha=0.5) # флотация сырьё
    df_full[metal[2]].hist(bins=25, label='rougher.output.concentrate', alpha=0.5) # флотация концентрат
    df_full[metal[3]].hist(bins=25, label='primary_cleaner.output', alpha=0.5) # первый этап
    df_full[metal[4]].hist(bins=25, label='secondary_cleaner.output', alpha=0.5) # второй этап
    plt.title('Диаграмма концентрации металла ' + metal[0])
    
    # подписи
    plt.legend(loc=legend)
    plt.xlabel('Значения')
    plt.ylabel('Количество')
    
    plt.show()
    
    # концентрация металла в сырье
    print('Концентрация металла в сырье:\n', df_full[[metal[2], 
                                                      metal[3], 
                                                      metal[4]]].mean())
    # концентрация металла в хвостах
    print('\nКонцентрация металла в хвостах: \n', df_full[[metal[5], metal[6], metal[7]]].mean())
```

### Сравним распределения размеров гранул сырья на обучающей и тестовой выборках

Для этого используем график плотности `kdeplot()` из библиотеки **seaborn**. \
Код отрисовки графиков также поместим в функцию.

```python
def granule_density(stage, limits):
    # размер графика
    plt.figure(figsize=(15, 5))
    
    # график
    sns.kdeplot(df_train[stage], 
            label=f'train {stage}', alpha=0.5, shade=True)
    sns.kdeplot(df_test[stage], 
            label=f'test {stage}', alpha=0.7, shade=True)
    
    # доп. параметры
    plt.xlim(limits)
    plt.xlabel('Значения')
    plt.ylabel('Плотность вероятности непрерывной величины')
    plt.legend(loc='upper right')
    
    plt.show()
    
    # характеристика распределений
    print('df_train:')
    print(df_train[stage].describe())
    
    print('\ndf_test:')
    print(df_test[stage].describe())
```

```python
# размер гранул сырья на входе флотации
granule_density('rougher.input.feed_size', [0, 175])
```

```python
# размер гранул сырья на входе финальной очистки
granule_density('primary_cleaner.input.feed_size', [5, 10])
```

### Напишем функцию для вычисления итоговой sMAPE

Для решения задачи введём новую метрику качества — **sMAPE** (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»). \
Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.

Метрика sMAPE вычисляется так:


```python
# sMAPE для одного этапа
def sMAPE(target, pred):
    
    smape = abs(target - pred) / ((abs(target) + abs(pred)) / 2) * 100
    
    #smape = smape.fillna(value=0)
    
    # складываем и делим на N
    smape = np.mean(smape)
    
    return smape
```

```python
# вычисление итоговой sMAPE, когда на входе target_rouger, pred_rougher, target_final, pred_final
def summary_smape_func(target_rouger, pred_rougher, target_final, pred_final):
    
    # вычислим smape для rouger
    smape_rouger = sMAPE(target_rouger, pred_rougher)
    
    # вычислим smape для final
    smape_final = sMAPE(target_final, pred_final)
    
    #  вычислим итоговый smape
    summary_smape = 0.25 * smape_rouger + 0.75 * smape_final
    
    return summary_smape
```

```python
# вычисления итоговой sMAPE, когда на входе smape_rouger и smape_final
def summary_smape(smape_rouger, smape_final):
    
    #  вычислим итоговый smape
    summary_smape = 0.25 * smape_rouger + 0.75 * smape_final
    
    return summary_smape
```
